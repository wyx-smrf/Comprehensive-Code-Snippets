# Categorical Scatterplot

Scatter plot separated by a categorical variable arranged in rows.

---

```python
scat_params = {'data':               df,
               'x_axis':            'num_colname1',
               'columns':           'cat_colname1',
               'y_axis':            'num_colname2',
               'plot_sep':          'cat_colname2'}

def cat_scatter(data, rows, columns, y_axis, plot_sep):
    scatter_plot = sns.FacetGrid(data, margin_titles=True,
                                 hue=plot_sep, col=columns,
                                 palette={1:"seagreen", 0:"gray"})
    
    scatter_plot.map(plt.scatter, y_axis, rows, edgecolor="w").add_legend()
    
    return plt.show()
    
cat_scatter(**scat_params)
```

## Example
---

```python
scat_params = {'data':              titanic,
               'rows':              'Age',
               'columns':           'Pclass',
               'y_axis':            'Fare',
               'plot_sep':          'Survived'}
```

```python
scat_param2 = {'data':              titanic,
               'x_axis':            'Fare',
               'columns':           'Sex',
               'y_axis':            'Age',
               'plot_sep':          'Survived'}
```

```python
def cat_scatter(data, columns, x_axis, y_axis, plot_sep):
    scatter_plot = sns.FacetGrid(titanic, 
                   hue = plot_sep,
                   col = columns, 
                   margin_titles=True,
                   palette="Set1",
                   hue_kws=dict(marker=["^", "v"]))
    
    scatter_plot.map(plt.scatter, x_axis, y_axis,edgecolor="w").add_legend()
    plt.subplots_adjust(top=0.8)
    scatter_plot.fig.suptitle('Survival by Gender , Age and Fare');

    return plt.show()

cat_scatter(**scat_param2)
```
